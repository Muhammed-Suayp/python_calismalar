```python
import pandas as pd

# Dosya yolu
dosya_yolu = r'C:\Users\suayp\Desktop\AI\suicide.csv'

def veri_incele(dosya_yolu):
    try:
        # Dosyayı oku
        dataset = pd.read_csv(dosya_yolu)
        print("Dosya başarıyla okundu.")
    except FileNotFoundError:
        print(f"{dosya_yolu} bulunamadı.")
        return
    except Exception as e:
        print(f"Dosya okunurken bir hata oluştu: {str(e)}")
        return
    
    # İlk ve son beş satırı yazdır
    print("\nİlk beş satır:")
    print(dataset.head())
    print("\nSon beş satır:")
    print(dataset.tail())
    
    # Temel istatistikleri yazdır
    print("\nTemel İstatistikler:")
    print(dataset.describe())
    
    # Eksik verileri kontrol et
    print("\nEksik Veri Kontrolü:")
    print(dataset.isnull().sum())

    # Sütun isimlerini yazdır
    print("\nSütun İsimleri:")
    print(dataset.columns)

# Fonksiyonu çağır
veri_incele(dosya_yolu)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dosya yolu
dosya_yolu = r'C:\Users\suayp\Desktop\AI\suicide.csv'

# Dosyayı oku
try:
    dataset = pd.read_csv(dosya_yolu)
    print("Dosya başarıyla okundu.")
except Exception as e:
    print(f"Dosya okunurken bir hata oluştu: {str(e)}")
    exit()

# Veri dağılımını anlamak için histogramlar çiz
dataset.hist(figsize=(10, 10))
plt.show()

# Korelasyon matrisi oluştur ve ısı haritası çiz
correlation_matrix = dataset.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True)
plt.show()

# Kategorik verileri say ve çubuk grafikleri çiz
for column in dataset.select_dtypes(include=['object']).columns:
    dataset[column].value_counts().plot(kind='bar', title=column)
    plt.show()
```
