```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python

plt.figure(figsize = (20,6))
area = [48,30,20,15]
priority = ['Low', 'Medium', 'High', 'Critical']
status = ['Resolved', 'Cancelled', 'Pending', 'Assigned']
company = ['IBM', 'Tesla', 'Microsoft', 'Apple']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']


fig, axs = plt.subplots(2, 3)


axs[0, 0].pie(area, labels = priority, colors = colors, startangle  = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
axs[0, 0].add_artist(my_circle)


axs[0, 1].pie(area, labels = status, colors = colors, startangle  = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
axs[0, 1].add_artist(my_circle)


axs[0, 2].pie(area, labels = company, colors = colors, startangle  = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
axs[0, 2].add_artist(my_circle)

axs[1, 0].pie(area, labels = company, colors = colors, startangle  = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
axs[1, 0].add_artist(my_circle)

axs[1, 1].pie(area, labels = company, colors = colors, startangle  = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
axs[1, 1].add_artist(my_circle)

axs[1, 2].pie(area, labels = company, colors = colors, startangle  = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
axs[1, 2].add_artist(my_circle)

plt.show()
```
